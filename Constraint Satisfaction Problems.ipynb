<h3>5.2 CSP Class</h3>
<p>Here we have the constraint satisfaction problem class which is going to have all the function we will use to solve our problem.</p>
<p>In this class we have variables and domains and constraints, then we check if every variable have a domain assigned to it or not.</p>
<p>Then we have three functions, the first one Add_constraint is to add the constraint we have to the constraints dictionary, the second method is consistent which is going to recieve a variable and checks it's constraints if they are consistent or not, the third method is isSatisfy this method is created to check if two variable satisfy the constraints or not, although this method is created the same in another class but it is created here to give easy accessibilty</p>

<p dir = "rtl">هنا لدينا CSP كلاس والتي ستمتلك جميع الوظائف التي سنستخدمها لحل مشكلتنا.

  لدينا متغيرات ومجالات وقيود ، نتحقق مما إذا كان لكل متغير مجال مخصص له أم لا.

 ثم لدينا ثلاث وظائف ، أولها Add_constraint هو إضافة القيد الذي لدينا إلى قاموس القيود ، والطريقة الثانية consistent والتي ستتلقى متغيرًا وتتحقق من قيودها إذا كانت متسقة أم لا ، والطريقة الثالثة هي isSatisfy  تم إنشاء هذه الطريقة للتحقق مما إذا كان هناك متغيران يفيان بالقيود أم لا ، على الرغم من أن هذه الطريقة يتم إنشاؤها بنفس الطريقة في فئة أخرى ، إلا أنها تم إنشاؤها هنا لإتاحة إمكانية الوصول بسهولة</p>

```python
class CSP(Generic[V, D]):
	def __init__(self, variables: List[V], domains: Dict[V, List[D]]) -> None:
		self.variables: List[V] = variables
		self.domains: Dict[V, List[D]] = domains
		self.constraints: Dict[V, List[Constraint[V, D]]] = {}

		for variable in self.variables:
			self.constraints[variable] = []
			if variable not in self.domains:
				raise LookupError("Every variable should have a domain assigned to it.")

	def add_constraint(self, constraint: Constraint[V, D]) -> None:
		for variable in constraint.variables:
			if variable not in self.variables:
				raise LookupError("Variable in constraint not in CSP")
			else:
				self.constraints[variable].append(constraint)

	def consistent(self, variable: V, assignment: Dict[V, D]) -> bool:
		for constraint in self.constraints[variable]:
			if not constraint.satisfied(assignment):
				return False
		return True


	def isSatisfy(self, pos1_value, pos2_value, pos1, pos2) -> bool:
		if pos1_value == pos2_value:
			return False
		if pos1 == "Forward Wall" and pos2 == "Back Wall" or pos1 == "Left Wall" and pos2 == "Right Wall":
			if pos1_value == "Bed":
				if pos2_value == "Air Conditioner":
					return True
			if pos1_value == "Wardrobe":
				if pos2_value == "Mirror":
					return True
			if pos1_value == "Air Conditioner":
				if pos2_value == "Bed":
					return True
			if pos1_value == "Mirror":
				if pos2_value == "Wardrobe":
					return True
		if pos1 == "Forward Wall" and pos2 == "Left Wall" or pos1 == "Forward Wall" and pos2 == "Right Wall":
			if pos1_value == "Bed":
				if pos2_value == "Air Conditioner":
					return False
			if pos1_value == "Wardrobe":
				if pos2_value == "Mirror":
					return False
			if pos1_value == "Air Conditioner":
				if pos2_value == "Bed":
					return False
			if pos1_value == "Mirror":
				if pos2_value == "Wardrobe":
					return False
		return True
```

<p>Here we have implemented the Backtracking search method which will search the problem recursively and return a result or if there is no solution the method will return none which indicate that the problem can never be solved.</p>

<p dir = "rtl">لقد قمنا هنا بتطبيق طريقة البحث Backtracking التي ستبحث في المشكلة بشكل متكرر وإرجاع نتيجة أو إذا لم يكن هناك حل ، سترجع قيمة بصيغة None</p>

```python
	def backtracking_search(self, assignment: Dict[V, D] = {}) -> Optional[Dict[V, D]]:
		startTime = time.perf_counter_ns()
		if len(assignment) == len(self.variables):
			return assignment
		unassigned: List[V] = [v for v in self.variables if v not in assignment]
		first: V = unassigned[0]
		global BackTrackTime
		for value in self.domains[first]:
			local_assignment = assignment.copy()
			local_assignment[first] = value
			if self.consistent(first, local_assignment):
				result: Optional[Dict[V, D]] = self.backtracking_search(local_assignment)
				if result is not None:
					return result
					endTime = time.process_time_ns()
					BackTrackTime = endTime - startTime
					print("BackTrack time in ns is =")
					print(BackTrackTime)
		endTime = time.perf_counter_ns()
		BackTrackTime = endTime - startTime
		print("BackTrack time in ns is =")
		print(BackTrackTime)

		return None
```

<p>Then we have the implementation of the AC3 method which will check the arc consistency be revising the constraints if the revision is false we remove from the domain the value that made the csp inconsistence.</p>


<p dir = "rtl">ثم لدينا تنفيذ طريقة AC3 التي ستتحقق من تناسق القوس ومراجعة القيود إذا كانت المراجعة خاطئة ، نزيل من المجال القيمة التي جعلت csp غير متناسق.</p>

```python
	def revise(self, pos1, pos2):
		revised = False
		pos1_domain = self.domains[pos1]
		pos2_domain = self.domains[pos2]
		for pos1_value in pos1_domain:
			satisfies = False
			for pos2_value in pos2_domain:
				if self.isSatisfy(pos1_value, pos2_value, pos1, pos2):
					satisfies = True
			if not satisfies:
				pos1_domain.remove(pos1_value)
				revised = True
				print("Removed From Domain " + pos1)

		self.domains[pos1] = pos1_domain
		return revised

	def ac3(self, constraints):
		arcs = []
		constraintsList = constraints["Forward Wall"] + constraints["Left Wall"] + constraints["Back Wall"] + constraints["Right Wall"]
		for variableA in self.domains:
			for variableB in self.domains:
				for i in range(len(constraintsList)):
					if constraintsList[i].pos1 == variableA and constraintsList[i].pos2 == variableB:
						arcs.append(tuple((variableA, variableB)))
		queue = arcs[:]
		revised = False
		global AC3Time
		startTime = time.perf_counter_ns()
		while queue:
			(x, y) = queue.pop(0)
			revised = self.revise(x, y)
			if revised:
				neighbors = [neighbor for neighbor in arcs if neighbor[1] == x]
				queue = queue + neighbors
		ns = False
		for variable in self.domains:
			if variable is None:
				ns = True
		if (ns):
			print(
				"There is no solution, it is inconsistence")
		endTime = time.perf_counter_ns()
		AC3Time = endTime - startTime
		print("AC3 Time in ns is =")
		print(AC3Time)
		return not ns
```

<h3>5.3 roomConstraint Class.</h3>
<p>The room constraint class is the class which will be the constraint we will create, this class will have the domains and two variable which are position one and position two which represent the the two walls that we will have a constraint on them.</p>
<p>Then we have the satisfied method which is overrided from the constraint class, this method will take an assignment dictionary that has the assigned values and check if they satisfy the constraint the method will return true otherwise false.</p>

<p dir = "rtl">كلاس شروط الغرفه هو الكلاس الذي سننشئه ، وسيحتوي هذا الكلاس على المجالات و على متغيران هما الموضع الأول والموضع الثاني اللذان يمثلان الجدارين اللذين سنضع قيدًا عليهما.

 ثم لدينا الطريقة المحققه التي تم اعادة كتابتها من كلاس القيد ، ستأخذ هذه الطريقة قاموس تعيين يحتوي على القيم المخصصة وتتحقق مما إذا كانت تفي بالقيد ، ستعيد الطريقة صحيحة وإلا خاطئة.</p>

```python
class roomConstraint(Constraint[str, str]):
	def __init__(self, pos1: str, pos2: str) -> None:
		super().__init__([pos1, pos2])
		self.pos1: str = pos1
		self.pos2: str = pos2
		self.domains: Dict[str, List[str]] = domains
            
	def satisfied(self, assignment: Dict[str, str]) -> bool:
		if self.pos1 not in assignment or self.pos2 not in assignment:
			return True
		if assignment[self.pos1] == assignment[self.pos2]:
			return False
		if self.pos1 == "Forward Wall" and self.pos2 == "Back Wall" or self.pos1 == "Left Wall" and self.pos2 == "Right Wall":
			if assignment[self.pos1] == "Bed":
				if assignment[self.pos2] == "Air Conditioner":
					return True
			if assignment[self.pos1] == "Wardrobe":
				if assignment[self.pos2] == "Mirror":
					return True
			if assignment[self.pos1] == "Air Conditioner":
				if assignment[self.pos2] == "Bed":
					return True
			if assignment[self.pos1] == "Mirror":
				if assignment[self.pos2] == "Wardrobe":
					return True
		if self.pos1 == "Forward Wall" and self.pos2 == "Left Wall" or self.pos1 == "Forward Wall" and self.pos2 == "Right Wall":
			if assignment[self.pos1] == "Bed":
				if assignment[self.pos2] == "Air Conditioner":
					return False
			if assignment[self.pos1] == "Wardrobe":
				if assignment[self.pos2] == "Mirror":
					return False
			if assignment[self.pos1] == "Air Conditioner":
				if assignment[self.pos2] == "Bed":
					return False
			if assignment[self.pos1] == "Mirror":
				if assignment[self.pos2] == "Wardrobe":
					return False
		return True
```

```python
from typing import Generic, TypeVar, Dict, List, Optional
from abc import ABC, abstractmethod
import time


global BackTrackTime
global BackTrackWithAC3Time
global AC3Time
global FCTime

V = TypeVar('V')
D = TypeVar('D')


class Constraint(Generic[V, D], ABC):
	def __init__(self, variables: List[V]) -> None:
		self.variables = variables

	@abstractmethod
	def satisfied(self, assignment: Dict[V, D]) -> bool:
		...

class CSP(Generic[V, D]):
	def __init__(self, variables: List[V], domains: Dict[V, List[D]]) -> None:
		self.variables: List[V] = variables
		self.domains: Dict[V, List[D]] = domains
		self.constraints: Dict[V, List[Constraint[V, D]]] = {}

		for variable in self.variables:
			self.constraints[variable] = []
			if variable not in self.domains:
				raise LookupError("Every variable should have a domain assigned to it.")

	def add_constraint(self, constraint: Constraint[V, D]) -> None:
		for variable in constraint.variables:
			if variable not in self.variables:
				raise LookupError("Variable in constraint not in CSP")
			else:
				self.constraints[variable].append(constraint)

	def consistent(self, variable: V, assignment: Dict[V, D]) -> bool:
		for constraint in self.constraints[variable]:
			if not constraint.satisfied(assignment):
				return False
		return True


	def isSatisfy(self, pos1_value, pos2_value, pos1, pos2) -> bool:
		if pos1_value == pos2_value:
			return False
		if pos1 == "Forward Wall" and pos2 == "Back Wall" or pos1 == "Left Wall" and pos2 == "Right Wall":
			if pos1_value == "Bed":
				if pos2_value == "Air Conditioner":
					return True
			if pos1_value == "Wardrobe":
				if pos2_value == "Mirror":
					return True
			if pos1_value == "Air Conditioner":
				if pos2_value == "Bed":
					return True
			if pos1_value == "Mirror":
				if pos2_value == "Wardrobe":
					return True
		if pos1 == "Forward Wall" and pos2 == "Left Wall" or pos1 == "Forward Wall" and pos2 == "Right Wall":
			if pos1_value == "Bed":
				if pos2_value == "Air Conditioner":
					return False
			if pos1_value == "Wardrobe":
				if pos2_value == "Mirror":
					return False
			if pos1_value == "Air Conditioner":
				if pos2_value == "Bed":
					return False
			if pos1_value == "Mirror":
				if pos2_value == "Wardrobe":
					return False
		return True

	def backtracking_search(self, assignment: Dict[V, D] = {}) -> Optional[Dict[V, D]]:
		startTime = time.perf_counter_ns()
		if len(assignment) == len(self.variables):
			return assignment
		unassigned: List[V] = [v for v in self.variables if v not in assignment]
		first: V = unassigned[0]
		global BackTrackTime
		for value in self.domains[first]:
			local_assignment = assignment.copy()
			local_assignment[first] = value
			if self.consistent(first, local_assignment):
				result: Optional[Dict[V, D]] = self.backtracking_search(local_assignment)
				if result is not None:
					return result
					endTime = time.process_time_ns()
					BackTrackTime = endTime - startTime
					print("BackTrack time in ns is =")
					print(BackTrackTime)
		endTime = time.perf_counter_ns()
		BackTrackTime = endTime - startTime
		print("BackTrack time in ns is =")
		print(BackTrackTime)

		return None

	def revise(self, pos1, pos2):
		revised = False
		# Get x and y domains
		pos1_domain = self.domains[pos1]
		pos2_domain = self.domains[pos2]
		for pos1_value in pos1_domain:
			satisfies = False
			for pos2_value in pos2_domain:
				if self.isSatisfy(pos1_value, pos2_value, pos1, pos2):
					satisfies = True
			if not satisfies:
				pos1_domain.remove(pos1_value)
				revised = True
				print("Removed From Domain " + pos1)

		self.domains[pos1] = pos1_domain
		return revised

	def ac3(self, constraints):
		arcs = []
		constraintsList = constraints["Forward Wall"] + constraints["Left Wall"] + constraints["Back Wall"] + constraints["Right Wall"]
		for variableA in self.domains:
			for variableB in self.domains:
				for i in range(len(constraintsList)):
					if constraintsList[i].pos1 == variableA and constraintsList[i].pos2 == variableB:
						arcs.append(tuple((variableA, variableB)))
		queue = arcs[:]
		revised = False
		global AC3Time
		startTime = time.perf_counter_ns()
		while queue:
			(x, y) = queue.pop(0)
			revised = self.revise(x, y)
			if revised:
				neighbors = [neighbor for neighbor in arcs if neighbor[1] == x]
				queue = queue + neighbors
		ns = False
		for variable in self.domains:
			if variable is None:
				ns = True
		if (ns):
			print(
				"There is no solution, it is inconsistence")
		endTime = time.perf_counter_ns()
		AC3Time = endTime - startTime
		print("AC3 Time in ns is =")
		print(AC3Time)
		return not ns

	def BacktrackWithAc3(self, constraints):
		const = self.ac3(constraints)
		Bcheck = False
		for variable in self.domains:
			if len(self.domains[variable]) != 1:
				Bcheck = True

		global BackTrackWithAC3Time
		startTime = time.perf_counter_ns()
		if const:
			if Bcheck:
				print("After AC3 with BackTracking")
				print(self.backtracking_search())
		endTime = time.perf_counter_ns()
		BackTrackWithAC3Time = endTime - startTime
		print("BackTrackWithAC3 time in ns is =")
		print(BackTrackWithAC3Time)


	def Forward_Checking(self):
		FilterCounter = 0
		global FCTime
		startTime = time.perf_counter_ns()
		for variable in variables:
			if self.Filtering(variable):
				FilterCounter = FilterCounter + 1
		if FilterCounter == 0:
			print("No Domains have been filtered")
		endTime = time.perf_counter_ns()
		FCTime = endTime - startTime
		print("Forwardchecking time in ns is =")
		print(FCTime)

	def Filtering(self, variable) -> bool:
		for value in domains[variable]:
			for uv in self.constraints[variable]:
				for uvalue in self.domains[uv.pos2]:
					if not self.isSatisfy(uvalue, value, variable, uv.pos2):
						self.domains[uv.pos2].remove(uvalue)
						return True


class roomConstraint(Constraint[str, str]):
	def __init__(self, pos1: str, pos2: str) -> None:
		super().__init__([pos1, pos2])
		self.pos1: str = pos1
		self.pos2: str = pos2
		self.domains: Dict[str, List[str]] = domains

	def satisfied(self, assignment: Dict[str, str]) -> bool:
		if self.pos1 not in assignment or self.pos2 not in assignment:
			return True
		if assignment[self.pos1] == assignment[self.pos2]:
			return False
		if self.pos1 == "Forward Wall" and self.pos2 == "Back Wall" or self.pos1 == "Left Wall" and self.pos2 == "Right Wall":
			if assignment[self.pos1] == "Bed":
				if assignment[self.pos2] == "Air Conditioner":
					return True
			if assignment[self.pos1] == "Wardrobe":
				if assignment[self.pos2] == "Mirror":
					return True
			if assignment[self.pos1] == "Air Conditioner":
				if assignment[self.pos2] == "Bed":
					return True
			if assignment[self.pos1] == "Mirror":
				if assignment[self.pos2] == "Wardrobe":
					return True
		if self.pos1 == "Forward Wall" and self.pos2 == "Left Wall" or self.pos1 == "Forward Wall" and self.pos2 == "Right Wall":
			if assignment[self.pos1] == "Bed":
				if assignment[self.pos2] == "Air Conditioner":
					return False
			if assignment[self.pos1] == "Wardrobe":
				if assignment[self.pos2] == "Mirror":
					return False
			if assignment[self.pos1] == "Air Conditioner":
				if assignment[self.pos2] == "Bed":
					return False
			if assignment[self.pos1] == "Mirror":
				if assignment[self.pos2] == "Wardrobe":
					return False
		return True



if __name__ == "__main__":
	variables: List[str] = ["Left Wall", "Forward Wall", "Back Wall", "Right Wall"]
	domains: Dict[str, List[str]] = {}
	for variable in variables:
		domains[variable] = ["Bed", "Air Conditioner", "Mirror", "Wardrobe"]

	csp: CSP[str, str] = CSP(variables, domains)
	csp.add_constraint(roomConstraint("Forward Wall", "Left Wall"))
	csp.add_constraint(roomConstraint("Forward Wall", "Back Wall"))
	csp.add_constraint(roomConstraint("Forward Wall", "Right Wall"))
	csp.add_constraint(roomConstraint("Left Wall", "Back Wall"))
	csp.add_constraint(roomConstraint("Left Wall", "Right Wall"))
	csp.add_constraint(roomConstraint("Right Wall", "Back Wall"))
```
